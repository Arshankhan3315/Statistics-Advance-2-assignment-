# Q2: What is Cumulative Density Function (CDF)? Explain with an example. Why CDF is used?

# Answer-2-you can calculate the Cumulative Density Function (CDF) for a given probability distribution using various libraries, such as SciPy. Let's demonstrate how to compute the CDF for a specific distribution using SciPy, specifically focusing on the normal distribution.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

mean = 0  
std_dev = 1  
x_values = np.linspace(-5, 5, 1000)
cdf_values = norm.cdf(x_values, mean, std_dev)
```

```python
plt.plot(x_values, cdf_values, label='CDF')
plt.xlabel('x')
plt.ylabel('CDF')
plt.title('Cumulative Density Function (CDF) of the Normal Distribution')
plt.legend()
plt.grid()
plt.show()
```

# Q5: What is Bernaulli Distribution? Give an Example. What is the difference between Bernoulli Distribution and Binomial Distribution?

# Answer-5-The Bernoulli distribution is a discrete probability distribution that models a single Bernoulli trial, which is a random experiment with exactly two possible outcomes: success or failure. It's named after the Swiss mathematician Jacob Bernoulli, who introduced it in the late 17th century.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import bernoulli, binom
```

```python
p_bernoulli = 0.3  
n_binomial = 10     
p_binomial = 0.3    

outcomes_bernoulli = np.array([0, 1])
successes_binomial = np.arange(0, n_binomial + 1)
pmf_bernoulli = bernoulli.pmf(outcomes_bernoulli, p_bernoulli)
pmf_binomial = binom.pmf(successes_binomial, n_binomial, p_binomial)
```

```python
plt.bar(outcomes_bernoulli, pmf_bernoulli, label='Bernoulli PMF')
plt.xlabel('Outcomes')
plt.ylabel('Probability')
plt.title('Bernoulli Distribution')
plt.xticks(outcomes_bernoulli)
plt.legend()
plt.show()
```

```python
plt.bar(successes_binomial, pmf_binomial, label='Binomial PMF')
plt.xlabel('Number of Successes')
plt.ylabel('Probability')
plt.title('Binomial Distribution')
plt.xticks(successes_binomial)
plt.legend()
plt.show()
```

# Q6. Consider a dataset with a mean of 50 and a standard deviation of 10. If we assume that the dataset is normally distributed, what is the probability that a randomly selected observation will be greater than 60? Use the appropriate formula and show your calculations.

# Answer-6-To calculate the probability that a randomly selected observation from a normally distributed dataset with a mean of 50 and a standard deviation of 10 is greater than 60, we'll use the cumulative distribution function (CDF) of the normal distribution. The formula for this is:

```python
import scipy.stats as stats
```

```python
mean = 50
std_dev = 10
value = 60
z_score = (value - mean) / std_dev
probability_less_than_60 = stats.norm.cdf(z_score)
probability_greater_than_60 = 1 - probability_less_than_60
```

```python
probability_greater_than_60
```

# Q7: Explain uniform Distribution with an example.

# Answer-7-The uniform distribution is a continuous probability distribution where every outcome within a specified range is equally likely to occur. In simple terms, it represents a situation where all values within a given interval have an equal probability of occurring.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
a = 0  
b = 10  
x_values = np.linspace(a - 2, b + 2, 400)  
pdf_values = np.full_like(x_values, 1 / (b - a))
```

```python
plt.plot(x_values, pdf_values, label='PDF')
plt.xlabel('x')
plt.ylabel('PDF')
plt.title('Uniform Distribution over [0, 10]')
plt.ylim(0, max(pdf_values) + 0.1)
plt.legend()
plt.grid()
plt.show()
```

# Q9: What is Central Limit Theorem? State the significance of the Central Limit Theorem.

# Answer-9-The Central Limit Theorem (CLT) is a fundamental concept in probability theory and statistics. It states that when independent random variables are added together, regardless of the original distribution of those variables, their sum tends to follow a normal distribution as the number of variables increases.

# Specifically, the Central Limit Theorem states that the sampling distribution of the sample mean (X) will be approximately normally distributed, regardless of the shape of the original population distribution, given a sufficiently large sample size. This is true even if the original population distribution is not normal.

# Significance of the Central Limit Theorem:

# Approximation to Normality:The Central Limit Theorem allows us to treat the sampling distribution of the sample mean as approximately normal, regardless of the underlying population distribution. This is critical in many statistical analyses.
# Inference and Hypothesis Testing:The CLT justifies the use of normal distribution-based statistical tests and confidence intervals in various real-world scenarios, making hypothesis testing and confidence interval estimation widely applicable.
# Large Sample Size Assumption:It provides a theoretical basis for the "large sample size" assumption in statistics. When the sample size is sufficiently large, we can apply methods that assume normality even if the population is not normally distributed.
# Estimation of Parameters:It allows for the estimation of population parameters like the mean and variance using the sample mean and variance, simplifying the estimation process.
# Foundation of Many Statistical Techniques:Many statistical techniques, including t-tests and ANOVA, rely on the Central Limit Theorem for their validity and applicability.
# In summary, the Central Limit Theorem is a crucial concept in statistics, providing a theoretical foundation for the application of many statistical methods, making statistical inference feasible and accurate in a wide range of practical situations.

# Q10: State the assumptions of the Central Limit Theorem.

# Answer-10-The Central Limit Theorem (CLT) is a fundamental theorem in statistics that outlines the conditions under which the sampling distribution of the sample mean approaches a normal distribution. However, it's important to note that the CLT holds under certain assumptions. Here are the key assumptions of the Central Limit Theorem:

# Independence of Observations:The observations in the sample should be independent. The values should not be influenced by each other.
# Identical Distribution (or Similar):The random variables should be drawn from the same probability distribution (or distributions that are very similar). This is often expressed as having the same mean (μ) and variance (2σ 2).
# Finite Variance:the random variables should have a finite variance (2σ 2).
# Random Sampling:The sample is obtained through a random sampling process, ensuring that each observation has an equal chance of being selected.
# Sufficiently Large Sample Size:The sample size (n) should be sufficiently large. While there is no strict threshold, larger sample sizes generally lead to a more accurate approximation of a normal distribution for the sample mean.
# Finite Population or Sampling with Replacement:If the sample is drawn from a finite population, the sample size should be less than 10% of the population, or the sample should be drawn with replacement to maintain approximate independence between observations.
# No Extreme Outliers:The presence of extreme outliers or very skewed distributions in the sample can affect the convergence to a normal distribution.
